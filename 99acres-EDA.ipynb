```python
#Importing required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
```

```python
# Importing dataset
df = pd.read_csv('data_for_99acres.com.csv')
```

```python
df.head(20)
```

```python
# Short Information about data
df.info()
```

```python
# Spliting the data based on space
new = df['apt_rent_month'].str.split(" ", n = 2, expand = True)

new.head()
```

```python
new.head(2)
```

```python
# Assign new name
new.columns = ['special_char','monthly_rant', 'Deposit']
```

```python
# Drop special char
new = new.drop('special_char',axis='columns')
new.head(2)
```

```python
# Removing Comma 
new["monthly_rant"]=new["monthly_rant"].str.replace(',', '')
```

```python
# Split the sentence to get only Deposit numbers
alpha = new['Deposit'].str.split('Deposit ', n = 5, expand = True)
alpha
```

```python
alpha.columns = ['garbage', 'Deposite']
```

```python
new = pd.merge(new,alpha,right_index=True,left_index=True)
new.head(2)
```

```python
# Removing Special Charater
new["Deposite"]=new["Deposite"].str.replace('₹', '')
```

```python
#Droping Unwanted column
new = new.drop(['Deposit','garbage'], axis=1)
```

```python
new.head(2) 
```

```python
new.monthly_rant.value_counts()
```

```python
# Merging the columns
df = pd.merge(df,new,right_index=True,left_index=True)
df.head(2)
```

```python
# Droped unwanted column
df = df.drop('apt_rent_month',axis=1)
df.head(2)
```

```python

new3 = df['description'].str.split(" ", n = 3, expand = True)
new3.head(5)
```

```python
new3.columns = ['no_of_BHK','garbage', 'Type_of_property', 'location_of_the_property']
```

```python
new3.head()
```

```python
new3 = new3.drop(['no_of_BHK','garbage'], axis='columns')
```

```python
new3.head(2)
```

```python
new3['location_of_the_property'] = new3['location_of_the_property'].str.split('in ', n = 3, expand = True)[1]
```

```python
new3.head()
```

```python
df = pd.merge(df,new3,right_index=True,left_index=True)
df.head(2)
```

```python
df = df.drop(['sq_ft','description'], axis=1)
```

```python
df.head(20)
```

```python
# Assigning on to a NaN value
df['monthly_rant']=df['monthly_rant'].replace('on','NaN')
```

```python
# Converting to a object to numeric
df['monthly_rant'] = pd.to_numeric(df['monthly_rant'], errors='coerce')
```

```python
# Filtering the data
df[df['monthly_rant']<=61]
#df1 = [x for x in df['monthly_rant'] if x >= 61]*10000
```

```python
# Change the value of lakhs in thousands
df['monthly_rant']= df['monthly_rant'].apply(lambda x:x*100000 if x<100 and x>0.5 else x)
```

```python
df['monthly_rant']
```

```python
df.iloc[99934]
```

```python
df.isnull().sum()
```

```python
# Spliting based on "m"
nw=df['Deposite'].str.split(" m", n = 3, expand = True)
df['deposit']=nw[0]
df['deposit']=pd.to_numeric(df['deposit'], errors='coerce')

# Multiply feature monthly_rant to deposite where deposite value 0 to 12
for col,row in df.iterrows():
    if (row['deposit']<12 and row['deposit'] >=0 ):
        df.loc[col,"deposit"]= row["deposit"] * row["monthly_rant"]
    else:
        df.loc[col,"deposit"]= row["deposit"]
```

```python
df.drop('Deposite',axis=1 , inplace=True)
```

```python
df.head(2)
```

```python
df.to_csv('99acres_cleaned.csv',index=False)
```
